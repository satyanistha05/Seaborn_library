```python
import pandas as pd
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Dataset.csv")
df
```

```python
sb.lineplot(data=df,x='PassengerId',y='Age')
plt.show()
```

```python
sb.lineplot(data=df,x='PassengerId',y='Age',hue='Sex')
plt.show()
```

```python
sb.lineplot(data=df,x='PassengerId',y='Age',hue='Embarked')
plt.show()
```

```python
sb.scatterplot(data=df,x='PassengerId',y='Age',hue='Embarked')
plt.show()
```

```python
sb.scatterplot(data=df,x='PassengerId',y='Age',hue='Sex')
plt.show()
```

```python
sb.scatterplot(data=df,x='PassengerId',y='Age')
plt.show()
```

```python
sb.boxplot(data=df,y='Age')
plt.show()
```

```python
sb.boxplot(data=df,y='Age',x='Sex',hue='Embarked')
plt.show()
```

```python
sb.violinplot(data=df,y='Age',x='Sex',hue='Embarked')
plt.show()
```

```python
sb.swarmplot(data=df,y='Age')
plt.show()
```

```python
sb.swarmplot(data=df,y='Age',x='Sex',hue='Embarked')
plt.show()
```

```python
sb.countplot(data=df,x='Sex')
plt.show()
```

```python
sb.countplot(data=df,x='Sex',hue='Pclass')
plt.show()
```

```python
tc=df[['Age','Fare','PassengerId']].corr()
sb.heatmap(tc,annot=True,fmt='0.3f')
```

```python
sb.displot(data=df,x='Age',hue="Pclass",kde=True)
plt.show()
```

```python
sb.kdeplot(data=df,x='Age',hue="Pclass")
plt.show()
```
